```python
from astropy.io import fits
import aplpy
import matplotlib.pyplot as plt
import numpy as np
```

## Following fits file is taken from https://skyview.gsfc.nasa.gov/current/cgi/query.pl with coordinate:
## 161.265, -59.685

```python
###This file is read using fits.open
with fits.open('skv9767250763118_4.fits') as hdu:
    data=hdu[0].data
    header=hdu[0].header
```

```python
##Checking the info of file
hdu.info()
```

```python
##Checking the shape of the data
print(data.shape)
```

## Plotting using imshow

```python
##plotting fits file
plt.figure(figsize=(10,8))
ax=plt.imshow(data,cmap='viridis',vmin=1.77490234375, vmax=8413.1953125)
plt.gca().invert_yaxis()
plt.colorbar(ax)
```

## WE can make similar plot even in better manner using aplpy

```python
fig=plt.figure(figsize=(10,10))
f=aplpy.FITSFigure('skv9767250763118_4.fits',figure=fig)
f.show_colorscale(cmap='viridis')
f.add_grid()
f.grid.set_linewidth(1.0)
f.grid.set_linestyle('dashed')
f.grid.set_alpha(0.3)
f.grid.show()
```
